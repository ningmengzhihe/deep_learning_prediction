# Save to Json

```python
import json
import pickle
```

```python
def save_to_json(pkl_filename, json_filename):
#     predictions = pickle.load(open('./results_save/predictions_for_plot.pkl', 'rb'))
    predictions = pickle.load(open(f'./results_save/{pkl_filename}.pkl', 'rb'))
    
    pre_list = [{"x":i, "y":predictions[i].item()} for i in range(0,predictions.shape[0])]
    ground_list = [{"x":i, "y":y_test[i].item()} for i in range(0,y_test.shape[0])]
    results = {}
    results["Predictions"] = pre_list
    results["GroundTruth"] = ground_list
    
    json.dump(results, open(f'/Users/lihan/Workspace/GULP_Project/data/{json_filename}.json', 'w'))
```

# Extract results and save to json

```python
import json
import pickle
```

```python
def save_to_json_10times(pkl_filename, json_filename):
    results_10times = pickle.load(open(f'./results_save/deep_learning_results_10times/{pkl_filename}.pkl', 'rb'))
    predictions = results_10times['y_test_pre']
    
    pre_list = [{"x":i, "y":predictions[i].item()} for i in range(0,predictions.shape[0])]
    ground_list = [{"x":i, "y":y_test[i].item()} for i in range(0,y_test.shape[0])]
    results = {}
    results["Predictions"] = pre_list
    results["GroundTruth"] = ground_list
    
    json.dump(results, open(f'/Users/lihan/Workspace/GULP_Project/data/{json_filename}.json', 'w'))
    
    # print metrics
    print('mse:', results_10times['mse'])
    print('rmse:', results_10times['rmse'])
    print('mae:', results_10times['mae'])
    print('r2:', results_10times['r2'])
```
